2.   **What is an interpreter in Python?**

The interpreter (CPython, PyPy, etc.) reads your Python code and executes it line by line, translating it to machine-level actions at runtime.

3. **What are pre-defined keywords in Python?**

Reserved words that have special meaning and can’t be used as identifiers (e.g., if, else, for, while, def, class, True, False, None, and, or, not, import, from, try, except, return, etc.).

4.**Can keywords be used as variable names?**

5.**What is mutability in Python?**

Mutability means an object can change after creation. Mutable examples: list, dict, set. Immutable examples: int, float, str, tuple, frozenset.

6.**Why are lists mutable, but tuples are immutable?**

By design: lists are meant for collections you modify (append, remove, sort). Tuples are fixed sequences—safer as dictionary keys and for protecting data from accidental changes, and they can be stored more compactly/hashed.

7.**Difference between == and is python operators?**

 == checks value equality (do they look the same?).
 is checks identity (are they the exact same object in memory?).

```python
a = [1,2]
b = [1,2]
a == b # True (same contents)
a is b # False ( different objects)
```

8. What are logical operators in Python?
and, or, not. They short-circuit and return one of the operands:

```python
True and False # False
True or False # True
not True  # False
```

9. **What is type casting in Python?**

Converting a value from one type to another, e.g., int("42"), float(5), str(3.14), list("hi") -> ['h','i'].

10. **What is the difference between implicit and explicit type casting?**

Implicit: Python automatically converts when safe (e.g., int + float -> float).

Explicit: You convert manually using constructors (int(), float(), str(), etc.).

11. **What is the purpose of conditional statements in Python?**

#13.Write a python program to find the sum of all even numbers between 1 and 50

```python
total = 0

for i in range(1, 51):
    if i % 2 == 0:
        total += i

print("Sum of even numbers between 1 and 50 is:", total)
```

#14.Write a python program to reverese a string using a while loop

```python
text = input("Enter a string: ")

i = len(text) - 1

while i >= 0:
    print(text[i], end="")
    i -= 1
```

#15. Write a python program to calculate the factoria of a number provided by user using a while loop

```python
num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
